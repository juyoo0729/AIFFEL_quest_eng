````python
# %% [markdown]
# # 한국어 → 영어 번역기 v2 (수정판)
#
# **이전 버전과의 차이점:**
# 1. ✨ **Repetition Penalty** - "obama obama obama..." 반복 문제 해결
# 2. ✨ **N-gram Blocking** - 같은 구절 반복 방지
# 3. ✨ **에폭 25로 증가** - 더 충분한 학습 (15 → 25)
# 4. ✨ **체크포인트 자동 저장** - 매 에폭마다 저장, 끊겨도 이어서 학습 가능
# 5. ✨ **번역 비교 출력** - 페널티 적용 전 vs 후 비교
#
# **작업 폴더**: C:\AI_study\ko-us\
# **데이터**: korean-english-park.train.tar.gz
# **환경**: Windows + RTX 3070


# %% 1. 패키지 설치 (최초 1회만)
# !pip install kiwipiepy torch matplotlib tqdm pandas numpy


# %% 2. 라이브러리 임포트 + 환경 설정
import os
import re
import sys
import math
import time
import random
import tarfile
import pickle
from pathlib import Path
from collections import Counter

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
from tqdm import tqdm

import matplotlib.pyplot as plt
import platform

from kiwipiepy import Kiwi

# 한글 폰트 (OS 자동 감지)
if platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
elif platform.system() == "Darwin":
    plt.rcParams["font.family"] = "AppleGothic"
else:
    plt.rcParams["font.family"] = "NanumGothic"
plt.rcParams["axes.unicode_minus"] = False

# 시드 고정
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"디바이스: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")


# %% 3. 데이터 압축 해제 + 로딩
DATA_DIR = Path(".")
TAR_PATH = DATA_DIR / "korean-english-park.train.tar.gz"

if not TAR_PATH.exists():
    raise FileNotFoundError(f"데이터 파일 없음: {TAR_PATH.absolute()}")

# 이미 압축 해제했으면 재해제 안함
ko_files = list(DATA_DIR.glob("*.ko"))
en_files = list(DATA_DIR.glob("*.en"))

if not ko_files or not en_files:
    print("압축 해제 중...")
    with tarfile.open(TAR_PATH, "r:gz") as tar:
        tar.extractall(DATA_DIR)
    ko_files = list(DATA_DIR.glob("*.ko"))
    en_files = list(DATA_DIR.glob("*.en"))

KO_FILE = ko_files[0]
EN_FILE = en_files[0]

with open(KO_FILE, "r", encoding="utf-8") as f:
    raw_ko = f.read().strip().split("\n")
with open(EN_FILE, "r", encoding="utf-8") as f:
    raw_en = f.read().strip().split("\n")

print(f"한국어 문장: {len(raw_ko):,}")
print(f"영어 문장: {len(raw_en):,}")
assert len(raw_ko) == len(raw_en)


# %% 4. 중복 제거
print(f"중복 제거 전: {len(raw_ko):,}쌍")
pair_set = set(zip(raw_ko, raw_en))
print(f"중복 제거 후: {len(pair_set):,}쌍")
cleaned_corpus = list(pair_set)


# %% 5. 정규식 전처리
def preprocess_korean(sentence):
    sentence = sentence.lower().strip()
    sentence = re.sub(r"([?.!,])", r" \1 ", sentence)
    sentence = re.sub(r"[^가-힣a-zA-Z0-9?.!,]+", " ", sentence)
    sentence = re.sub(r"\s+", " ", sentence)
    return sentence.strip()


def preprocess_english(sentence):
    sentence = sentence.lower().strip()
    sentence = re.sub(r"([?.!,])", r" \1 ", sentence)
    sentence = re.sub(r"[^a-zA-Z0-9?.!,]+", " ", sentence)
    sentence = re.sub(r"\s+", " ", sentence)
    return sentence.strip()


print("정규식 전처리 중...")
preprocessed = []
for ko, en in tqdm(cleaned_corpus, desc="전처리"):
    ko_clean = preprocess_korean(ko)
    en_clean = preprocess_english(en)
    if ko_clean and en_clean:
        preprocessed.append((ko_clean, en_clean))

print(f"전처리 후: {len(preprocessed):,}쌍")


# %% 6. Kiwi 토큰화 함수
print("Kiwi 초기화 중...")
kiwi = Kiwi()
print("완료")


def tokenize_korean(sentence):
    return [token.form for token in kiwi.tokenize(sentence)]


def tokenize_english(sentence):
    return ["<start>"] + sentence.split() + ["<end>"]


# %% 7. 토큰화 + 길이 40 이하 필터링
MAX_TOKEN_LEN = 40

print("토큰화 + 길이 필터링...")
ko_corpus = []
en_corpus = []

for ko, en in tqdm(preprocessed, desc="토큰화"):
    ko_tokens = tokenize_korean(ko)
    en_tokens = tokenize_english(en)
    if len(ko_tokens) <= MAX_TOKEN_LEN and len(en_tokens) <= MAX_TOKEN_LEN:
        ko_corpus.append(ko_tokens)
        en_corpus.append(en_tokens)

print(f"길이 {MAX_TOKEN_LEN} 이하: {len(ko_corpus):,}쌍")


# %% 8. Vocabulary 생성
PAD_IDX, START_IDX, END_IDX, UNK_IDX = 0, 1, 2, 3
SPECIAL_TOKENS = ["<pad>", "<start>", "<end>", "<unk>"]

VOCAB_SIZE_KO = 15000
VOCAB_SIZE_EN = 12000


def build_vocab(token_lists, vocab_size):
    counter = Counter()
    for tokens in token_lists:
        counter.update(tokens)
    word2idx = {tok: i for i, tok in enumerate(SPECIAL_TOKENS)}
    most_common = counter.most_common(vocab_size - len(SPECIAL_TOKENS))
    for word, _ in most_common:
        if word not in word2idx:
            word2idx[word] = len(word2idx)
    idx2word = {i: w for w, i in word2idx.items()}
    return word2idx, idx2word


ko_word2idx, ko_idx2word = build_vocab(ko_corpus, VOCAB_SIZE_KO)
en_word2idx, en_idx2word = build_vocab(en_corpus, VOCAB_SIZE_EN)
print(f"한국어 vocab: {len(ko_word2idx):,}, 영어 vocab: {len(en_word2idx):,}")


def encode(tokens, word2idx):
    return [word2idx.get(tok, UNK_IDX) for tok in tokens]


def decode(ids, idx2word):
    return [idx2word.get(int(i), "<unk>") for i in ids]


# %% 9. Dataset / DataLoader
class TranslationDataset(Dataset):
    def __init__(self, ko_corpus, en_corpus, ko_w2i, en_w2i):
        self.ko_corpus = ko_corpus
        self.en_corpus = en_corpus
        self.ko_w2i = ko_w2i
        self.en_w2i = en_w2i
    
    def __len__(self):
        return len(self.ko_corpus)
    
    def __getitem__(self, idx):
        ko_ids = encode(self.ko_corpus[idx], self.ko_w2i)
        en_ids = encode(self.en_corpus[idx], self.en_w2i)
        return torch.tensor(ko_ids, dtype=torch.long), torch.tensor(en_ids, dtype=torch.long)


def collate_fn(batch):
    src_seqs, tgt_seqs = zip(*batch)
    src_padded = pad_sequence(src_seqs, batch_first=True, padding_value=PAD_IDX)
    tgt_padded = pad_sequence(tgt_seqs, batch_first=True, padding_value=PAD_IDX)
    return src_padded, tgt_padded


BATCH_SIZE = 64
dataset = TranslationDataset(ko_corpus, en_corpus, ko_word2idx, en_word2idx)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, 
                    collate_fn=collate_fn, num_workers=0)


# %% 10. Bahdanau Attention
class BahdanauAttention(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.W1 = nn.Linear(hidden_dim, hidden_dim)
        self.W2 = nn.Linear(hidden_dim, hidden_dim)
        self.v = nn.Linear(hidden_dim, 1, bias=False)
    
    def forward(self, hidden, encoder_outputs, src_mask):
        src_len = encoder_outputs.size(0)
        hidden = hidden.unsqueeze(0).repeat(src_len, 1, 1)
        energy = torch.tanh(self.W1(encoder_outputs) + self.W2(hidden))
        scores = self.v(energy).squeeze(2).transpose(0, 1)
        scores = scores.masked_fill(src_mask == 0, -1e9)
        attention_weights = F.softmax(scores, dim=1)
        encoder_outputs_t = encoder_outputs.transpose(0, 1)
        context = torch.bmm(attention_weights.unsqueeze(1), encoder_outputs_t).squeeze(1)
        return context, attention_weights


# %% 11. Encoder
class Encoder(nn.Module):
    def __init__(self, input_dim, emb_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(input_dim, emb_dim, padding_idx=PAD_IDX)
        self.rnn = nn.GRU(emb_dim, hidden_dim)
    
    def forward(self, src):
        embedded = self.embedding(src)
        outputs, hidden = self.rnn(embedded)
        return outputs, hidden


# %% 12. Decoder (with Attention)
class Decoder(nn.Module):
    def __init__(self, output_dim, emb_dim, hidden_dim, attention):
        super().__init__()
        self.output_dim = output_dim
        self.attention = attention
        self.embedding = nn.Embedding(output_dim, emb_dim, padding_idx=PAD_IDX)
        self.rnn = nn.GRU(emb_dim + hidden_dim, hidden_dim)
        self.fc = nn.Linear(hidden_dim + hidden_dim + emb_dim, output_dim)
    
    def forward(self, input_tok, hidden, encoder_outputs, src_mask):
        input_tok = input_tok.unsqueeze(0)
        embedded = self.embedding(input_tok)
        context, attn_weights = self.attention(hidden.squeeze(0), encoder_outputs, src_mask)
        rnn_input = torch.cat([embedded, context.unsqueeze(0)], dim=2)
        output, hidden = self.rnn(rnn_input, hidden)
        output = output.squeeze(0)
        embedded = embedded.squeeze(0)
        prediction = self.fc(torch.cat([output, context, embedded], dim=1))
        return prediction, hidden, attn_weights


# %% 13. Seq2Seq with Attention
class Seq2SeqAttention(nn.Module):
    def __init__(self, encoder, decoder, device):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device
    
    def forward(self, src, tgt=None, max_len=40, teacher_forcing_ratio=0.5):
        batch_size = src.size(1)
        src_len = src.size(0)
        src_mask = (src.transpose(0, 1) != PAD_IDX).long()
        encoder_outputs, hidden = self.encoder(src)
        
        if tgt is not None:
            tgt_len = tgt.size(0)
            outputs = torch.zeros(tgt_len, batch_size, self.decoder.output_dim).to(self.device)
            input_tok = tgt[0]
            for t in range(1, tgt_len):
                output, hidden, _ = self.decoder(input_tok, hidden, encoder_outputs, src_mask)
                outputs[t] = output
                use_tf = random.random() < teacher_forcing_ratio
                top1 = output.argmax(1)
                input_tok = tgt[t] if use_tf else top1
            return outputs
        else:
            outputs = torch.zeros(max_len, batch_size, self.decoder.output_dim).to(self.device)
            attentions = torch.zeros(max_len, batch_size, src_len).to(self.device)
            input_tok = torch.tensor([START_IDX] * batch_size, dtype=torch.long).to(self.device)
            for t in range(max_len):
                output, hidden, attn = self.decoder(input_tok, hidden, encoder_outputs, src_mask)
                outputs[t] = output
                attentions[t] = attn
                input_tok = output.argmax(1)
                if (input_tok == END_IDX).all():
                    break
            return outputs, attentions


# %% 14. 모델 인스턴스화
input_dim = len(ko_word2idx)
output_dim = len(en_word2idx)
EMB_DIM = 256
HID_DIM = 512

encoder = Encoder(input_dim, EMB_DIM, HID_DIM).to(device)
attention = BahdanauAttention(HID_DIM).to(device)
decoder = Decoder(output_dim, EMB_DIM, HID_DIM, attention).to(device)
model = Seq2SeqAttention(encoder, decoder, device).to(device)

n_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"모델 파라미터: {n_params:,}")


# %% 15. 학습 준비 + 체크포인트 함수
"""
★ 새로 추가: 매 에폭마다 체크포인트 저장
중간에 학습 중단되면 마지막 체크포인트에서 이어 가능
"""
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)

CHECKPOINT_DIR = Path("checkpoints")
CHECKPOINT_DIR.mkdir(exist_ok=True)


def save_checkpoint(epoch, model, optimizer, losses, path):
    """체크포인트 저장"""
    torch.save({
        "epoch": epoch,
        "model_state": model.state_dict(),
        "optimizer_state": optimizer.state_dict(),
        "losses": losses,
        "config": {
            "input_dim": input_dim, "output_dim": output_dim,
            "emb_dim": EMB_DIM, "hid_dim": HID_DIM,
        },
    }, path)


def load_checkpoint(path, model, optimizer):
    """체크포인트 불러오기"""
    if not path.exists():
        return 0, []
    ckpt = torch.load(path, map_location=device)
    model.load_state_dict(ckpt["model_state"])
    optimizer.load_state_dict(ckpt["optimizer_state"])
    print(f"[체크포인트 로드] {ckpt['epoch']}에폭부터 이어 학습")
    return ckpt["epoch"], ckpt["losses"]


def train_step(model, loader, optimizer, criterion, epoch, total_epochs):
    model.train()
    epoch_loss = 0
    progress_bar = tqdm(loader, desc=f"Epoch {epoch+1}/{total_epochs}", leave=True)
    
    for src, tgt in progress_bar:
        src = src.permute(1, 0).to(device)
        tgt = tgt.permute(1, 0).to(device)
        
        optimizer.zero_grad()
        outputs = model(src, tgt, teacher_forcing_ratio=0.5)
        outputs_flat = outputs[1:].reshape(-1, outputs.size(-1))
        tgt_flat = tgt[1:].reshape(-1)
        loss = criterion(outputs_flat, tgt_flat)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        
        epoch_loss += loss.item()
        progress_bar.set_postfix(loss=f"{loss.item():.4f}")
    
    return epoch_loss / len(loader)


# %% 16. 학습 실행 (★ 수정: 25 에폭 + 매 에폭 체크포인트 저장)
"""
RTX 3070 기준 25 에폭에 약 30~40분.
끊겨도 마지막 체크포인트부터 이어서 학습 가능.

처음 학습: RESUME = False
이어서 학습: RESUME = True
"""
EPOCHS = 25
RESUME = False  # ← 처음 학습이면 False, 이어서 하면 True

CHECKPOINT_LATEST = CHECKPOINT_DIR / "latest.pt"

if RESUME:
    start_epoch, losses = load_checkpoint(CHECKPOINT_LATEST, model, optimizer)
else:
    start_epoch, losses = 0, []

print(f"\n학습 시작 (에폭 {start_epoch+1}~{EPOCHS})")
print("=" * 60)

start_time = time.time()

for epoch in range(start_epoch, EPOCHS):
    loss = train_step(model, loader, optimizer, criterion, epoch, EPOCHS)
    losses.append(loss)
    
    # ★ 매 에폭마다 체크포인트 저장
    save_checkpoint(epoch + 1, model, optimizer, losses, CHECKPOINT_LATEST)
    
    elapsed = time.time() - start_time
    print(f"Epoch {epoch+1:2d}/{EPOCHS} | Loss: {loss:.4f} | "
          f"PPL: {math.exp(loss):.2f} | 경과: {elapsed/60:.1f}분")

print("=" * 60)
print(f"학습 완료. 총 {(time.time()-start_time)/60:.1f}분")


# %% 17. ★ 핵심 추가: Repetition Penalty + N-gram Blocking 번역 함수
"""
이전 결과에서 "obama obama obama..." 같은 반복 문제가 있었음.
이를 해결하기 위한 두 가지 기법 추가:

1. Repetition Penalty: 이미 생성한 단어는 다음에 덜 뽑히게 logit 깎음
2. N-gram Blocking: 같은 N-gram이 두 번 안 나오게 차단
"""

def translate(sentence, model, ko_w2i, en_i2w, max_len=40,
              repetition_penalty=2.0, no_repeat_ngram=3):
    """
    repetition_penalty: 1.0=페널티 없음, 2.0=권장, 클수록 강함
    no_repeat_ngram: 같은 N개 단어 묶음이 두 번 나오지 않도록 차단 (3 권장)
    """
    model.eval()
    
    with torch.no_grad():
        # 전처리 + 토큰화
        sentence_clean = preprocess_korean(sentence)
        ko_tokens = tokenize_korean(sentence_clean)
        ko_ids = encode(ko_tokens, ko_w2i)
        src = torch.tensor(ko_ids, dtype=torch.long).unsqueeze(1).to(device)
        src_mask = (src.transpose(0, 1) != PAD_IDX).long()
        
        # Encoder
        encoder_outputs, hidden = model.encoder(src)
        input_tok = torch.tensor([START_IDX], dtype=torch.long).to(device)
        
        pred_tokens = []
        attentions_list = []
        generated_ids_seq = []  # 순서대로 생성된 토큰 ID
        
        for t in range(max_len):
            output, hidden, attn = model.decoder(input_tok, hidden, encoder_outputs, src_mask)
            
            # ★ 1. Repetition Penalty
            if repetition_penalty > 1.0:
                for prev_id in set(generated_ids_seq):
                    if output[0, prev_id] > 0:
                        output[0, prev_id] /= repetition_penalty
                    else:
                        output[0, prev_id] *= repetition_penalty
            
            # ★ 2. N-gram Blocking
            if no_repeat_ngram > 0 and len(generated_ids_seq) >= no_repeat_ngram - 1:
                recent = tuple(generated_ids_seq[-(no_repeat_ngram - 1):])
                
                banned_ids = set()
                for i in range(len(generated_ids_seq) - no_repeat_ngram + 1):
                    ngram_prefix = tuple(generated_ids_seq[i:i + no_repeat_ngram - 1])
                    if ngram_prefix == recent:
                        banned_ids.add(generated_ids_seq[i + no_repeat_ngram - 1])
                
                for bid in banned_ids:
                    output[0, bid] = -1e9
            
            pred_id = output.argmax(1).item()
            
            if pred_id == END_IDX:
                pred_tokens.append("<end>")
                attentions_list.append(attn.squeeze(0).cpu().numpy())
                break
            
            pred_tokens.append(en_i2w.get(pred_id, "<unk>"))
            generated_ids_seq.append(pred_id)
            attentions_list.append(attn.squeeze(0).cpu().numpy())
            
            input_tok = torch.tensor([pred_id], dtype=torch.long).to(device)
        
        attn_arr = np.array(attentions_list) if attentions_list else np.zeros((1, src.size(0)))
        return pred_tokens, ko_tokens, attn_arr


def plot_attention(attention, src_tokens, pred_tokens, title=""):
    fig, ax = plt.subplots(figsize=(10, 8))
    cax = ax.matshow(attention, cmap="viridis", aspect="auto")
    fig.colorbar(cax, ax=ax)
    ax.set_xticks(range(len(src_tokens)))
    ax.set_xticklabels(src_tokens, rotation=90, fontsize=11)
    ax.set_yticks(range(len(pred_tokens)))
    ax.set_yticklabels(pred_tokens, fontsize=11)
    ax.set_xlabel("Source (Korean)")
    ax.set_ylabel("Predicted (English)")
    if title:
        ax.set_title(title, pad=20)
    plt.tight_layout()
    return fig


# %% 18. ★ 비교 출력: 페널티 적용 전 vs 후
"""
페널티 효과를 직접 비교해서 확인!
"""
test_sentences = [
    "오바마는 대통령이다.",
    "시민들은 도시 속에 산다.",
    "커피는 필요 없다.",
    "일곱 명의 사망자가 발생했다.",
]

print("=" * 70)
print("번역 결과 비교: 페널티 적용 전 vs 후")
print("=" * 70)

results_with_penalty = []

for i, sent in enumerate(test_sentences, 1):
    # 페널티 없이 (이전 버전과 동일)
    pred_no, _, _ = translate(sent, model, ko_word2idx, en_idx2word,
                               repetition_penalty=1.0, no_repeat_ngram=0)
    
    # 페널티 적용
    pred_yes, ko_tokens, attn = translate(sent, model, ko_word2idx, en_idx2word,
                                           repetition_penalty=2.0, no_repeat_ngram=3)
    
    print(f"\n[K{i}] {sent}")
    print(f"  페널티 X: {' '.join(pred_no)}")
    print(f"  페널티 O: {' '.join(pred_yes)}")
    
    results_with_penalty.append({
        "k": sent,
        "e": " ".join(pred_yes),
        "ko_tokens": ko_tokens,
        "pred_tokens": pred_yes,
        "attention": attn,
    })


# %% 19. ★ 최종 제출용 결과 (페널티 적용)
print("\n" + "=" * 70)
print("【퀘스트 제출용】")
print("=" * 70)
for i, r in enumerate(results_with_penalty, 1):
    print(f"\nK{i}) {r['k']}")
    print(f"E{i}) {r['e']}")


# %% 20. Attention Map 시각화
for i, r in enumerate(results_with_penalty, 1):
    fig = plot_attention(
        r["attention"], r["ko_tokens"], r["pred_tokens"],
        title=f"K{i}: {r['k']} → {r['e'][:50]}..."
    )
    save_path = CHECKPOINT_DIR / f"attention_K{i}_v2.png"
    fig.savefig(save_path, dpi=100, bbox_inches="tight")
    plt.show()
    print(f"저장: {save_path}")


# %% 21. 학습 곡선
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(losses) + 1), losses, marker="o", linewidth=2)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title(f"Training Loss ({len(losses)} epochs)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(CHECKPOINT_DIR / "training_loss_v2.png", dpi=100)
plt.show()


# %% 22. 최종 모델 + Vocab 저장
final_model_path = CHECKPOINT_DIR / "ko_en_translator_v2.pt"
vocab_path = CHECKPOINT_DIR / "vocab_v2.pkl"

torch.save({
    "model_state": model.state_dict(),
    "config": {
        "input_dim": input_dim, "output_dim": output_dim,
        "emb_dim": EMB_DIM, "hid_dim": HID_DIM,
    },
    "losses": losses,
}, final_model_path)

with open(vocab_path, "wb") as f:
    pickle.dump({
        "ko_word2idx": ko_word2idx, "ko_idx2word": ko_idx2word,
        "en_word2idx": en_word2idx, "en_idx2word": en_idx2word,
    }, f)

print(f"모델 저장: {final_model_path}")
print(f"Vocab 저장: {vocab_path}")


# %% 23. 추가 번역 테스트
extra_tests = [
    "한국은 아름다운 나라이다.",
    "나는 오늘 영화를 봤다.",
    "정부는 새로운 정책을 발표했다.",
    "학생들이 시험을 준비하고 있다.",
]

print("=" * 70)
print("추가 번역 테스트 (페널티 적용)")
print("=" * 70)
for sent in extra_tests:
    pred, _, _ = translate(sent, model, ko_word2idx, en_idx2word,
                           repetition_penalty=2.0, no_repeat_ngram=3)
    print(f"\nKO) {sent}")
    print(f"EN) {' '.join(pred)}")


# %% [markdown]
# ## 사용 팁
#
# **페널티 강도 조절** (셀 18):
# - `repetition_penalty=1.0` → 페널티 없음 (이전 버전)
# - `repetition_penalty=1.5` → 약한 페널티
# - `repetition_penalty=2.0` → 권장 (이번 기본값)
# - `repetition_penalty=2.5` → 강한 페널티
#
# **N-gram 차단 강도** (셀 18):
# - `no_repeat_ngram=2` → 같은 2단어 묶음 차단 (강함)
# - `no_repeat_ngram=3` → 같은 3단어 묶음 차단 (권장)
# - `no_repeat_ngram=4` → 같은 4단어 묶음 차단 (느슨함)
# - `no_repeat_ngram=0` → 차단 없음
#
# **학습 끊겼을 때 이어서 하려면**:
# 셀 16의 `RESUME = True`로 변경하고 다시 실행
#
# **결과가 여전히 별로라면**:
# - EPOCHS를 25 → 30으로 더 늘리기
# - 셀 14의 HID_DIM을 512 → 768로 늘리기
````

## 어려웠던 점과 해결 과정

### 1. Repetition 문제 (핵심)

처음 학습 후 결과를 보고 충격받았다.

```
"obama obama obama obama obama president obama obama obama..."
"i i i i i i i i i i i i i i i i i i..."
"killed killed killed killed killed killed..."
```

학습이 잘못된 줄 알았는데 알고 보니 **RNN 기반 Seq2Seq의 고질적 문제**였다. Greedy decoding이라 한번 같은 단어를 뽑으면 무한 루프에 빠지는 것이다.

**해결 방법**: 추론 단계에서 두 가지 기법 추가
1. **Repetition Penalty**: 이미 생성한 단어는 다음 시점에서 logit을 깎아서 덜 뽑히게
2. **N-gram Blocking**: 같은 N개 단어 묶음이 다시 나오지 않도록 차단

```python
# Repetition Penalty 핵심 코드
if repetition_penalty > 1.0:
    for prev_id in set(generated_ids_seq):
        if output[0, prev_id] > 0:
            output[0, prev_id] /= repetition_penalty
        else:
            output[0, prev_id] *= repetition_penalty
```

K3의 경우 페널티 적용 전엔 `need`가 30번 넘게 반복됐는데, 적용 후엔 의미 있는 문장으로 바뀌었다.

→ **Kiwi**로 대체하고, OS 자동 감지 코드(`platform.system()`)로 폰트 설정도 통일했다.

## 배운 점

### 1. 모델 구조보다 데이터와 디코딩 전략이 중요할 수 있다

처음엔 모델만 잘 짜면 결과가 좋을 줄 알았다. 실제로는 **같은 모델이라도 추론 단계에서 어떻게 단어를 뽑는지**(Greedy vs Beam vs Sampling, 페널티 적용 여부)에 따라 결과가 크게 달라진다.

K4 결과만 봐도:
- 페널티 X → `killed`만 30번 반복하는 의미 없는 출력
- 페널티 O → `seven`, `deaths`, `dead`가 들어간 의미 있는 문장

같은 모델, 같은 가중치인데 결과가 완전히 다르다.

### 2. Attention의 위력

Attention Map을 시각화해보니 모델이 한국어 토큰 중 어디에 집중하는지가 시각적으로 보였다. **추상적이던 개념이 시각화로 와닿은 순간**이었다.

### 3. PPL이 낮다고 결과가 완벽하진 않다

PPL 6.75는 꽤 낮은 값이지만, 실제 번역 결과는 여전히 완벽하지 않다. **정량 지표와 실제 품질 간 갭**이 있다는 걸 알게 됐다. BLEU 같은 추가 평가 지표도 필요하다.

### 4. RNN의 한계

데이터를 늘리고 에폭을 키워도 RNN 기반 Seq2Seq는 한계가 있다. 긴 문장에서 정보 손실, 반복 문제, 학습 속도 등이 모두 RNN의 구조적 한계에서 온다. **다음 단계가 왜 Transformer인지** 직접 만들어보면서 절감했다.

## 마무리

전체적으로 결과가 완벽하진 않지만, NLP 흐름의 큰 그림(**Word2Vec → Seq2Seq → Attention → Transformer**)이 잡힌 프로젝트였다.

특히 **모델 학습보다 추론 단계의 디코딩 전략이 결과를 크게 좌우한다**는 걸 직접 경험한 게 가장 큰 수확이었다. 처음엔 결과가 너무 안 좋아서 모델 자체에 문제가 있는 줄 알았는데, 추론 함수만 손보니 결과가 확 좋아진 게 인상 깊었다.

다음에 Transformer로 같은 데이터를 학습해서 직접 비교해보고 싶다.
